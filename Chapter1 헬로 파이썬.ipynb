### 1.4.2 클래스

- int, str 등의 자료형은 파이썬이 기본으로 제공하는, 내장된 자료형

- 하지만 직접 클래스를 정의하여 클래스만의 전용 함수, 속성을 정의할 수 있다



```
class 클래스 이름:
    def __init__(self, 인수, ...):
        ...
    def 메서드 이름 1(self, 인수, ...):
        ...
    def 메서드 이름 2(self, 인수, ...):
        ...
```

- \__init__\: 클래스를 초기화하는 방법을 정의한다. 이 초기화용 메서드를 생성자라고도 하며, 클래스의 인스턴스가 만들어질 때 한번만 불린다

- 메서드의 첫번째 인수로 자신(자신의 인스턴스)을 나타내는 self를 명시적으로 쓰는 것이 특징이다




````python
class Man:
    def __init__(self, name): # 클래스의 속성 정의. 
        self.name = name
        print('Initialized!')

    def hello(self):
        print('Hello ' + self.name + '!')
    
    def goodbye(self):
        print('Good-bye ' + self.name + '!')

m = Man('John')
m.hello() # 클래스 안의 존재하는 함수를 선택해서 사용가능
m.goodbye()
````

- 위의 클래스를 man.py로 만들어서 저장하고 실행시킬 것이다.

````python
!python man.py
````

- Man이라는 새로운 클래스를 정의하고 / Man 클래스에서 m이라는 인스턴스(객체)를 생성한다

- Man의 생성자(초기화 메서드)는 name이라는 인수를 받고 / 그 인수로 인스턴수 변수인 self.name을 초기화 한다

- 인스턴스 변수는 인스턴스별로 저장되는 변수

- 파이썬에서는 self.name처럼 self다음에 속성 이름을 써서 인스턴스 변수를 작성하거나 접근할 수 있다

### 1.5.4 넘파이의 N차원 배열

### 1.5.5 브로드캐스트

- 넘파이에서 형상이 다른 배열끼리 계산가능

   ex) 2x2 행렬과 스칼라값 10의 곱 

   => 스칼라값 10이 2x2 행렬로 확대된 후에 연산이 이루어지는 것을 브로드캐스트라 한다



````python
A = np.array([[1,2], [3,4]])
B = np.array([[10, 20]])
A*B
````

### 1.5.6 원소 접근

- 리스트에서 인덱싱한 것과 비슷한데 차원이 더 늘어나서 인덱싱하는 자리가 늘어남

````python
X[0] # 0번째 행렬
````

````python
X[0][1] # 0번째 행렬의 1번째 열에 있는 값
````

````python
for row in X: # 행렬을 for문을 돌리면 행렬 하나씩 순서대로 들어간다
    print(row)
````

````python
X = X.flatten()
X, X.shape # 3x2 행렬에서 6,벡터로 변환
````

````python
X[[0,2,4]] # 원하는 인덱스를 꼭 넘파이 형태로 넣지 않아도 댐
````

````python
X > 15 # 넘파이 배열에 부등호 연산을 수행하면 배열의 원소 각각에 부등호 연산을 수행한 bool배열이 생성된다
````

````python
X[X>15] # 판다스 조건 넣는거랑 비슷한듯?
````

## Matplotlib

- 그래프를 그려줌으로 데이터의 시각화

### 1.6.1 단순한 그래프 그리기

- matplotlib의 pyplot 모듈을 이용해 그래프를 그린다

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1간격으로 생성
y = np.sin(x)

plt.plot(x,y)
plt.show()
````

### 1.6.2 pyplot의 기능

- 제목, 각 축의 이름(레이블) 표시 등...

````python
x = np.arange(0, 6, 0.1) 
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle='--', label='cos') # linestyle = '--', 점선
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')
plt.legend() # 점선과 선이 어떤걸 나타내는지 표시해줌
plt.show()
````
